```python
import xlrd
import pandas as pd
import numpy as np
import json
import unidecode
```

```python
df = pd.read_excel('input/youth_survey_raw.xlsx')

# get question ids from columns
question_ids = [ c for c in df.columns if 'Q' in c ]

# copy row 0 to column names
df.columns = df.iloc[0]

# delete row 0
df = df.drop(0, axis=0)
df = df.applymap(str)

df
```

```python
# compile list of question_columns - could be neater!
question_columns = []

for c in df.columns:
    for q_id in question_ids:
        if q_id in c:
            question_columns.append(c)
```

```python
# for each Governorate and District pair...
# show how many records in this pair
# for key, df_ in df.groupby( ['M2: Governorate', 'M3: District'] ):
#     print(key)
#     print( len(df_) )


```

```python
def getNational():
    
    overall = {}
    for q in question_columns:
#         print()
#         print (q)
        q_dict = {}
        values_list = []
        count_list = []
        
        unique, counts = np.unique(df[q].values, return_counts=True)
        
        for q_, c in zip(unique, counts):
            if q_ != 'nan':
                
                count_list.append(c)
            
                answer = unidecode.unidecode(q_.replace("(DK)", "Don't know"))
                answer = answer.replace('3.', '')

                values = {
                    "answer": answer,
                    "count": c
                }
                
                values_list.append(values)
        
        a_dict = {}
        for item in values_list:
            total = sum(count_list) 
            item["pct"] = float((item["count"]/total)*100)
            item["count"] = int(item["count"])
            a = item["answer"]
            
            del item["answer"]

            
            
            a_dict[a] = item
            
            
#         print ('----')
#         print (a_dict)

        
        if len(q.split(':')) > 1:
            q = q.split(':')[1].strip()
            
        if q != "Other":
            if len(a_dict) > 0:
                overall[q] = a_dict
                
        
                
    return overall
        


nationalData = getNational()
nationalData
```

```python
with open('output/national.json', 'w') as outfile:
    json.dump(nationalData, outfile)
```
