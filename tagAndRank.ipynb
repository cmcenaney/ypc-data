```python
import xlrd
import pandas as pd
import numpy as np
import json
import unidecode
```

```python
df = pd.read_excel('input/youth_survey_raw.xlsx')

# get question ids from columns
question_ids = [ c for c in df.columns if 'Q' in c ]

# copy row 0 to column names
df.columns = df.iloc[0]

# delete row 0
df = df.drop(0, axis=0)
df = df.applymap(str)

df
```

```python
# compile list of question_columns - could be neater!
question_columns = []

for c in df.columns:
    for q_id in question_ids:
        if q_id in c:
            question_columns.append(c)
```

```python
df_rank = pd.DataFrame(columns=["question", "answer", "rank"])
df_tag = pd.DataFrame(columns=["question", "tag"])

q_rank_list = []
q_tag_list = []
a_list = []

for q in question_columns:
    unique, counts = np.unique(df[q].values, return_counts=True)
    
    if len(q.split(':')) > 1:
        q = q.split(':')[1].strip()
        q = unidecode.unidecode(q)
    
    if q != "Other":
        
        if q not in q_tag_list:
            q_tag_list.append(q)
        
        for a, c in zip(unique, counts):
            a = unidecode.unidecode(a.replace("(DK)", "Don't know")).replace("3.", "")
            if a != "nan":
                if a != "(RF)":

    #                 if len(q.split(':')) > 1:
    #                     q = q.split(':')[1].strip()

#                     if q != "Other":
                    if q not in q_rank_list:
                        q_rank_list.append(q)
            
                        a_list.append(a)
                
                        print ()
                        print (q)

df_rank["question"] = q_rank_list
df_rank["answer"] = a_list
df_tag["question"] = q_tag_list

df_rank = df_rank.fillna("")
df_tag = df_tag.fillna("")

df_question_rename = df_tag.rename(columns = {"question": "question", "tag": "question_display"})
df_answer_rename = df_rank.rename(columns = {"question": "question", "answer": "answer", "rank": "answer_display"})



writer = pd.ExcelWriter('ypc-tag-rank-rewrite.xlsx')
df_tag.to_excel(writer,'tag_questions', index=False)
df_question_rename.to_excel(writer,'rewrite_questions', index=False)
df_rank.to_excel(writer,'rank_answers', index=False)
df_answer_rename.to_excel(writer, 'rewrite_answers', index=False)
writer.save()
```
